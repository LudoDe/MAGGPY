## âœ… Verify the installation

Run the cell below to import every critical package. If nothing errors out, you're good to go!

````python
import importlib, sys

packages = {
    "numpy":      "numpy",
    "scipy":      "scipy",
    "astropy":    "astropy",
    "pandas":     "pandas",
    "matplotlib": "matplotlib",
    "h5py":       "h5py",
    "healpy":     "healpy",
    "emcee":      "emcee",
    "corner":     "corner",
    "joblib":     "joblib",
    "tqdm":       "tqdm",
    "GWFish":     "GWFish",
    "lal":        "lal (lalsuite)",
    "gdt.core":   "astro-gdt (core)",
    "gdt.missions.fermi.gbm": "astro-gdt-fermi",
}

print(f"Python {sys.version}\n")
all_ok = True
for module, label in packages.items():
    try:
        mod = importlib.import_module(module)
        ver = getattr(mod, "__version__", "â€”")
        print(f"  âœ…  {label:.<30s} {ver}")
    except ImportError:
        print(f"  âŒ  {label:.<30s} NOT FOUND")
        all_ok = False

print()
if all_ok:
    print("ğŸ‰ All packages installed successfully â€” you're ready for the tutorials!")
else:
    print("âš ï¸  Some packages are missing. Re-run the cells above or check the error messages.")
````
